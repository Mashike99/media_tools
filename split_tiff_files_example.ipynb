## Multipage TIFF Splitter

Recently our lab has been doing a lot of analysis and measurement of citrus tree trunk xylem vessels in images we take on our Nikon Eclipse Ti2 microscope. Most of the routine measurements can be automated fairly easily using macros in the [Fiji](https://imagej.net/software/fiji/) distribution of ImageJ. However others have to be processed using image analysis modules in Python or R. When importing an image sequence into ImageJ, the resulting image analysis files (in this case, black circles on a white background) are often exported to the multipage .tiff format for ease. You can easily split the file into individual images using ImageJ, but when dealing with large tiff stacks the file saving process can be a bit cumbersome. Also most .tiff splitting tools online have filesize limits, which is a problem when dealing with microscopic images

I wanted to write some code to automate the process for us and export the resulting files to a directory of choice. The script `binary_tiff_splitter.py` contains a function `tiff_splitter` to help us out here.


```python
from binary_tiff_splitter import tiff_splitter

import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's take a look at the docstring for `tiff_splitter`.

```python
print(tiff_splitter.__doc__)
```

We can inspect the arguments that this function takes as well.

```python
import inspect
print(inspect.signature(tiff_splitter))
```

Let's start with a simple example to test out the code. In our directory, we have a file named `tree_trunk_xylem.tif` that is a binary .tif image with multiple pages. We will pass the filename as the first argument. The argument `target_dir` when equal to `None` (default value) will save the output files in your current directory. Let's set it to save in a subdirectory called `tif_output`. *Note: tiff_splitter will create the `target_dir` for you if it doesn't currently exist*

```python
tiff_splitter('tree_trunk_xylem.tif', 'tif_output')
```

Lets plot one of the images to show that we indeed split the tiff file.


```python
img1 = plt.imread('tif_output/img_1.tif')
plt.figure(figsize=(15, 15))
plt.imshow(img1, cmap = "gray", vmin = 0, vmax = 255)
```

By default `tiff_splitter` names the output files numerically prefixed by `img_`. You can specify your own output names by passing a list

```python
tiff_splitter('tree_trunk_xylem.tif', 'tif_output', output_names = ('test_1.tif', 'test_2.tif', 'test_3.tif'))
```

In some instances we may want to invert the images to make the black pixels white and white pixels black etc. *Note: this assumes that your image only has pixel values of 0 or 255 already. If your image is a color image, you'll need to use other image analysis tools to invert the pixel values*

The argument `invert_images` accepts boolean values.

```python
tiff_splitter('tree_trunk_xylem.tif', 
              'tif_output', 
              output_names = ('inv_1.tif', 'inv_2.tif', 'inv_3.tif'),
              invert_images=True)
```

```python
img1 = plt.imread('tif_output/inv_1.tif')
plt.figure(figsize=(15, 15))
plt.imshow(img1, cmap = "gray", vmin = 0, vmax = 255)
```

Finally, many times we will want to have access to the underlying numpy arrays of the data if we need to perform an operations on the matrices. The argument `return_data` if set to `True` will return an n-dimensional numpy array with the pixel values.

```python
arrays = tiff_splitter('tree_trunk_xylem.tif', 
              'tif_output', 
              output_names = ('inv_1.tif', 'inv_2.tif', 'inv_3.tif'),
              invert_images=True,
              return_data=True)
```

```python
print(arrays)
print(type(arrays))
```
